```python
# -*- coding: utf-8 -*-
import ipywidgets as w
from IPython.display import display, HTML

# ---------------- Datos ----------------
cursos = {
    'Ciclo 1': ['Cálculo diferencial e integral', 'Biologia (Bot)', 'Mecánica de fluidos y termodinámica', 'Introducción a la ingeniería ambiental','Química', 'Herramientas informáticas'],
    'Ciclo 2': ['Biología (Zoo)', 'Ambiente y Sociedad', 'Física Ambiental', 'Programación y base de datos', 'Ecuaciones diferenciales', '**'],
    'Ciclo 3': ['Microbiología Ambiental', 'Modelación Ambiental', 'Biofísica','Ecología', 'Estadística Descriptiva','(CN)Flora y fauna'],
    'Ciclo 4': ['Tecnologías de información geoespacial', 'Ecología SbN', 'Energías Renovables', 'Estadística Inferencial', '*'],
    'Ciclo 5': ['Química Ambiental', 'Hidrología y Meteorología', 'Economía Ambiental', 'Ecotoxicología','Educación Ambiental', 'Residuos Sólidos'],
    'Ciclo 6': ['Procesos Ambientales', 'Formulación de proyectos de sustentabilidad ambiental', 'Gestión Ambiental de Cuencas', 'Calidad del agua', 'Calidad del suelo', 'Calidad del aire'],
    'Ciclo 7': [ 'Metodología de la Investigación', 'Evaluación de Impactos Ambientales', 'Lesgislación, Regulación y Fiscalización Ambiental', 'Tecnologías para el tratamiento de aguas','Remediciación de suelo', 'Tratamiento de la contaminación del aire'],
    'Ciclo 8': ['Sistema de Gestión y Auditorías Ambientales', 'Gestión de residuos sólidos', 'Técnicas de Restauración Ambiental','Seguridad y Salud Ocupacional', 'Biotecnología y Remediación Ambiental', 'Formulación del proyecto de titulación'],
    'Ciclo 9': ['TIC', 'Humanismo y ética', 'Solución de conflictos ambientales', 'Comunicación Oral y Escrita'],
}

# ---------------- Controles ----------------
buscador = w.Text(placeholder='Filtrar asignaturas...', layout=w.Layout(width='320px'))
orden = w.ToggleButtons(options=[('Original', 'orig'), ('Alfabético', 'abc')], value='orig')
controles = w.HBox([buscador, orden])

contenedor = w.HBox(layout=w.Layout(overflow='auto', width='100%', height='420px', padding='4px'))

# ---------------- Estilos ----------------
display(HTML("""
<style>
.box-ciclo {border:1px solid #ddd;border-radius:10px;padding:8px;margin:6px;min-width:260px;max-width:260px;}
.titulo-ciclo {font-weight:700;margin-bottom:6px;text-align:center;}
.legend {font-size:12px;color:#666;margin-bottom:8px;text-align:center;}
.item {padding:2px 0;font-size:14px;border-bottom:1px dashed #eee;}
.item:last-child{border-bottom:none;}
.vacante {font-style:italic;color:#888;}
.badge {display:inline-block;padding:1px 6px;border:1px solid #bbb;border-radius:8px;font-size:11px;margin-left:6px;color:#555;}
</style>
"""))

# ---------------- Helpers ----------------
def columna_ciclo(nombre_ciclo, asignaturas):
    html = f"<div class='box-ciclo'><div class='titulo-ciclo'>{nombre_ciclo}</div>"
    html += "<div class='legend'>* = vacante para nueva asignatura</div>"
    if not asignaturas:
        html += "<div class='item' style='color:#888;'>Sin asignaturas</div>"
    else:
        for a in asignaturas:
            if a.strip() == '*':
                html += "<div class='item vacante'>• * <span class='badge'>vacante</span></div>"
            else:
                html += f"<div class='item'>• {a}</div>"
    html += "</div>"
    return w.HTML(value=html)

def _rebuild_with_star(original_list, items_sorted, star_pos):
    if star_pos is None:
        return items_sorted
    star_pos = max(0, min(star_pos, len(items_sorted)))
    rebuilt, i_items = [], 0
    for i in range(len(items_sorted) + 1):
        if i == star_pos:
            rebuilt.append('*')
        else:
            rebuilt.append(items_sorted[i_items])
            i_items += 1
    return rebuilt

def _order_preserve_star(lst, mode, query):
    star_pos = None
    for i, x in enumerate(lst):
        if x.strip() == '*':
            star_pos = i
            break
    non_star = [x for x in lst if x.strip() != '*']
    if query:
        non_star = [x for x in non_star if query in x.lower()]
    if mode == 'abc':
        non_star = sorted(non_star, key=lambda x: x.lower())
    return _rebuild_with_star(lst, non_star, star_pos)

# ---------------- Render ----------------
def render(_=None):
    q = buscador.value.strip().lower()
    columnas = []
    for ciclo, lista in cursos.items():
        lst = _order_preserve_star(lista, orden.value, q)
        columnas.append(columna_ciclo(ciclo, lst))
    contenedor.children = columnas

buscador.observe(render, names='value')
orden.observe(render, names='value')

render()
display(controles, contenedor)
```
